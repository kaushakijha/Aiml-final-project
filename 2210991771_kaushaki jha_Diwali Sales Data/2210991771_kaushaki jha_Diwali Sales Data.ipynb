# GRAPH TO SHOW THE MAXIMUM SALE GENDER WISE

```python
salesgra=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',hue='Amount',data=salesgra)
```

# MAXIMUM NUMBER OF ORDER PLACED

```python
sns.set(rc={'figure.figsize':(25,5)})
salesgra=df.groupby(['Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='Gender',y='Orders',hue='Gender',data=salesgra)
```

#CONCLUSION : MOST OF THE BUYERS ARE FEMALE

# STATE ANALYSIS

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='State',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salestate = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State',y='Orders',hue='Orders',data=salestate)
```

```python
salesamount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State' ,y='Amount',hue='Amount',data=salesamount)
```

CONCLUSION : FROM THE GRAPS WE SEE THAT MOST OF THE ORDERS AND SALES ARE FROM UP,MAHARASTRA AND KARNATAKA

# OCCUPATION ANALYSIS

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='Occupation',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df,hue='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue='Amount')
```

CONCLUSION : MOST OF THE SALE BUYERS ARE WORKING IN IT,HEALTHCARE AND AVIATION SECTOR

# PRODUCT CATEGORY ANALYSIS

```python
sns.set(rc={'figure.figsize':(30,15)})
ax = sns.countplot(data = df, x = 'Product_Category' ,hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount' ,hue='Amount')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue='Orders')
```

# conclusion : MARRIED WOMEN BETWEEN AGE GROUP 26-35 YEARS FROM UP ,MAHARASTRA AND KARNATAKE WORKING IN IT ,HEALTHCARE AND AVIATION ARE THE ONE WHO SHOPS THE MOST FROM FOOD CLOTHING AND ELECTRONIC CATEGORY
